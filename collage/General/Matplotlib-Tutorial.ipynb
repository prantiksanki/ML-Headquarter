```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.plot([1,2,3,4] , [2,4,6,9] , label= '2X' , color = 'red' , linewidth = 2 , marker = '.' , linestyle= "--" ,markersize = 10 , markeredgecolor = 'blue')
plt.title("Our First Graph" , fontdict ={'fontname': 'Comic Sans MS', 'fontsize' : 15 })
plt.xlabel("X Axis") 
plt.ylabel("Y Axis")
plt.xticks([0,1,2,3,4,5,6,7,8])
plt.yticks([0,2,4,6,8,10,12,14,16])
x2 = np.arange(0,4.5,0.5)
plt.plot(x2 , x2**2 , label = "x^2")
plt.figure(figsize = (2,1.5) , dpi = 300)
plt.savefig("file_name.png" , dpi = 300)
```

## BAR CHART

```python
labels = ['A' , 'B' , 'C']
values = [1,4,2] 

bars = plt.bar(labels, values)
plt.figure(figsize = (6,4))
bars[0].set_hatch('/')
bars[1].set_hatch('.')
bars[2].set_hatch('o')
```

## Real World Examples

```python
df = pd.read_csv('gas_prices.csv')
```

```python
df
```

```python
yr = df['Year'] 
can = df['Canada'] 
usa = df['USA'] 

plt.plot(yr,can , label="Canada" , color = 'green' , marker =".")
plt.plot(yr,usa, label ="USA" , color = 'red' , marker = ".")
plt.xlabel("Year") 
plt.ylabel("USD")
plt.title("Gas Price Over Time")
plt.legend()
plt.xticks(df.Year[::3])
plt.savefig("gas_price.png" , dpi = 300)
```

```python
for i in df :
    if i != 'Year':
        plt.plot(df.Year , df[i] , marker = "." )
plt.xlabel("Year") 
plt.ylabel("USD")
plt.title("Gas Price Over Time")
plt.xticks(df.Year[::3])
```
