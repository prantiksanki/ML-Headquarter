```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import RobustScaler
```

```python
df = pd.read_csv("employee_data.csv")
```

```python
df
```

```python
df1 = df.dropna()
```

```python
df1
```

```python
df
```

```python
df = df.fillna(round(df["Age"].mean(numeric_only=True)))
```

```python
df
```

```python
df = df.fillna(round(df["Bonus"].mean(numeric_only=True)))
```

```python
df
```

```python
columns_to_select = ["Age", "Salary", "Experience_Years", "Bonus"]
df_num = df[columns_to_select]
```

```python
df_num
```

```python
df_num
```

```python
#Standardization 
standard_scaler = StandardScaler()
df_standard = pd.DataFrame(standard_scaler.fit_transform(df_num))
```

```python
df_standard
```

```python
# Min-Max Scaling
min_max_scaler=MinMaxScaler()
df_min_max=pd.DataFrame(min_max_scaler.fit_transform(df_num),columns=df_num.columns)
```

```python
df_min_max
```

```python
# Robust Scalling
robust_scaler = RobustScaler()
df_robust = pd.DataFrame(
    robust_scaler.fit_transform(df_num),
    columns=df_num.columns  # Correct attribute
)
```

```python
df_robust
```
