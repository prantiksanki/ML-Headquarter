<p style="font-size: 18px;"> <b> 4. Univariate Analysis : </b> Plot the distribution of house prices. Is the data skewed? If so, what kind of transformation
might normalize it?</p> 

```python
# Plot the distribution of house prices
sns.histplot(df['Y house price of unit area'], bins=30, color='#40376E')
plt.title("House Price Plot Distribution")
plt.xlabel("House Price Per Unit Area")
plt.ylabel("Frequency")
```

```python
# Is the data skewed? 
skewness = df['Y house price of unit area'].skew()
skewness
```

<pre>As we know ,
Near 0: Symmetric distribution.
0: Right-skewed.
< 0: Left-skewed.
</pre>
<p> So it has symmetric distribution</p>

<p> <b>Normalization Technique : </b> <br>i) log transformation.  <br>ii) Square root transformation. </p>

```python
# i) Log transformation
np.log(df['Y house price of unit area'])
```

```python
# Square root transformation
np.sqrt(df['Y house price of unit area'])
```

<p style = "font-size:18px"> <b>5. Feature Exploration: </b>
Investigate the relationship between each available feature (X1 to X6) and house prices.
Use a scatter plot to visualize the relationship and interpret the results.<p>

```python
df
```

```python
indexes = ['X1 transaction date', 'X2 house age', 'X3 distance to the nearest MRT station', 'X4 number of convenience stores', 'X5 latitude', 'X6 longitude']
target = 'Y house price of unit area'
plt.figure(figsize=(15, 15))
for i, index in enumerate(indexes, 1):
    plt.subplot(2, 3, i)
    sns.scatterplot(x=df[index], y=df[target])
    plt.title(f"{index} vs {target}")
    plt.xlabel(index)
    plt.ylabel(target)
```
