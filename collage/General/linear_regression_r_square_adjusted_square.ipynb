```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("student_data.csv")
```

```python
data
```

```python
X = data.iloc[ : , : -1] 
y = data.iloc[ : , -1]
```

```python
X
```

```python
y
```

```python
model = LinearRegression() 
model.fit(X,y)
```

```python
y_pred = model.predict(X) 
```

```python
def calculate_r(y_true,_pred):
    correlation_matrix = np.corrcoef(y_true,_pred)
    r = correlation_matrix[0,1]
    return r
```

```python
def calculate_r2(y_true, y_pred):
    ss_total = np .sum (( y_true - np . mean ( y_true ) ) ** 2)
    ss_residual = np .sum (( y_true - y_pred ) ** 2)
    r2 = 1 - ( ss_residual / ss_total )
    return r2
```

```python
def calculate_adjusted_r2 ( y_true , y_pred , n_features ) :
    n_samples = len( y_true )
    r2 = calculate_r2 ( y_true , y_pred )
    adjusted_r2 = 1 - (1 - r2 ) * ( n_samples - 1) / (
        n_samples - n_features - 1)

    return adjusted_r2
```

```python
r = calculate_r (y , y_pred )
r2 = calculate_r2 (y , y_pred )
adjusted_r2 = calculate_adjusted_r2 (y , y_pred , X . shape [1])
print ( f" Correlation Coefficient (R): {r:.4f}")
print ( f" R Score : {r2 :.4f}")
print ( f" Adjusted R Score : { adjusted_r2 :.4f}")
```
