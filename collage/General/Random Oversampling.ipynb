```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from imblearn.over_sampling import RandomOverSampler
from collections import Counter
```

```python
# Step 1: Generate an imbalanced dataset
X, y = make_classification(n_samples=1000, n_features=2, n_informative=2, n_redundant=0, n_clusters_per_class=1, weights=[0.9, 0.1], flip_y=0, 
                           random_state=42)
```

```python

# Check class distribution
print("Original class distribution:", Counter(y))
```

```python

# Plot original dataset
plt.figure(figsize=(6, 4))
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], label="Majority Class (0)", alpha=0.6)
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], label="Minority Class (1)", alpha=0.6, color="red")
plt.title("Original Imbalanced Dataset")
plt.legend()
plt.show()
```

```python
# Step 2: Apply Random Over-Sampling
ros = RandomOverSampler(sampling_strategy='auto', random_state=42)
X_resampled, y_resampled = ros.fit_resample(X, y)

```

```python
# Check new class distribution
print("Balanced class distribution:", Counter(y_resampled))

```

```python
# Plot resampled dataset
plt.figure(figsize=(6, 4))
plt.scatter(X_resampled[y_resampled == 0][:, 0], X_resampled[y_resampled == 0][:, 1], 
            label="Majority Class (0)", alpha=0.6)
plt.scatter(X_resampled[y_resampled == 1][:, 0], X_resampled[y_resampled == 1][:, 1], 
            label="Minority Class (1)", alpha=0.6, color="red")
plt.title("Balanced Dataset After Random Over-Sampling")
plt.legend()
plt.show()
```
