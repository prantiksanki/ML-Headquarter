**1. Write a Python program to create a bar graph that displays the population of five
different cities. Customize the bar graph by adding a title, labels for the axes, and
different colors for each bar.**
___

```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
# Data
cities = ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix']
populations = [8419000, 3980000, 2716000, 2328000, 1690000]
#Code here :

df = pd.DataFrame({"City": cities, "Population": populations})
plt.bar(df["City"], df["Population"], color=["Red" , "Green", "Blue", "yellow", "Green"])
plt.xlabel("City")
plt.ylabel("Population")
plt.title("Population of Major U.S. Cities")
```

**2. Write a Python program to create a pie chart that represents the market share
 of different smartphone brands. Label each wedge with the brand name and percentage, and
 highlight (explode) the wedge representing higher market share.**
 ___

```python
# Data
brands = ['Apple', 'Samsung', 'Huawei', 'Xiaomi', 'Oppo']
market_share = [27.5, 21.3, 15.8, 10.5, 8.3]

#Code here::
max_index = market_share.index(max(market_share))
explode = [0.1 if i == max_index else 0 for i in range(len(brands))]
plt.figure(figsize=(8, 8))
plt.pie(market_share, labels=brands, autopct="%1.1f%%", startangle=140, explode=explode, colors=["gold", "lightblue", "red", "green", "purple", "gray"], shadow=True)
plt.title("Smartphone Market Share")
```

**5. Write a Python program to create a line plot that shows the performance of three different students over five exams.Plot all three lines on the same graph, using different colors and markers for each line. Add a legend to identify each student.**
___

```python
# Data
exams = [1, 2, 3, 4, 5]
student1_scores = [80, 85, 88, 90, 92]
student2_scores = [78, 82, 85, 87, 89]
student3_scores = [75, 80, 82, 85, 88]

#Code here::
plt.plot(exams, student1_scores, marker="o", linestyle="-", color="blue", label="Student A")
plt.plot(exams, student2_scores, marker="s", linestyle="-", color="red", label="Student B")
plt.plot(exams, student3_scores, marker="^", linestyle="-", color="green", label="Student C")

plt.xlabel("Exam Number")
plt.ylabel("Scores")
plt.title("Performance of Students")
```

**6. Write a Python program to create a stacked bar chart that represents the sales of three different products over four quarters. use different colors for each product, and add a legend.**
___

```python
# Data
quarters = ['Q1', 'Q2', 'Q3', 'Q4']
product1_sales = [150, 200, 250, 300]
product2_sales = [100, 150, 200, 250]
product3_sales = [50, 100, 150, 200]

#Code here::

plt.figure(figsize=(10, 6))
plt.bar(quarters, product1_sales, label='Product A', color='skyblue')
plt.bar(quarters, product2_sales, bottom=product1_sales, label='Product B', color='lightgreen')
plt.bar(quarters, product3_sales, bottom=np.array(product1_sales)+np.array(product2_sales), label='Product C', color='salmon')
plt.title('Product Sales by Quarter')
plt.xlabel('Quarters')
plt.ylabel('Sales')
```

**19. Using the Titanic dataset create a figure with two subplots: one subplot showing a histogram of the 'age' column with 50 bins and the second subplot showing a bar chart of the 'pclass' column. Ensure that both plots share the same x-axis.**
___

```python
#Code here:
fig, ax = plt.subplots(2, 1, figsize=(8, 10), sharex=True)

ax[0].hist(df["age"], bins=50, color="skyblue", edgecolor="black")
ax[0].set_title("Histogram of Age")
ax[0].set_ylabel("Frequency")

pclass_counts = df["pclass"].value_counts().sort_index() 
ax[1].bar(pclass_counts.index, pclass_counts.values, color="coral", edgecolor="black")
ax[1].set_title("Passenger Class Distribution")
ax[1].set_xlabel("Age / Passenger Class")  
ax[1].set_ylabel("Count")
ax[1].set_xticks(pclass_counts.index) 

plt.tight_layout()
```

**20. Using the Titanic dataset create a figure with two subplots: the first subplot should display a line plot of the 'age' column, and the second subplot should display a bar chart of the 'sex' column showing the count of each gender.**
___

```python
#Code here:
fig, ax = plt.subplots(2, 1, figsize=(8, 10))

ax[0].plot(df.index, df["age"], linestyle="-", color="blue", marker="o", markersize=3)
ax[0].set_title("Line Plot of Age")
ax[0].set_ylabel("Age")
ax[0].grid(True, linestyle=":", linewidth=0.7)

sex_counts = df["sex"].value_counts()
ax[1].bar(sex_counts.index, sex_counts.values, color=["lightblue", "pink"], edgecolor="black")
ax[1].set_title("Gender Distribution")
ax[1].set_xlabel("Sex")
ax[1].set_ylabel("Count")

plt.tight_layout()
```
