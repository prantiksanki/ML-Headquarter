<b>Fake News Prediction </b>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report, confusion_matrix

```

```python
true_news = pd.read_csv('True.csv')
fake_news = pd.read_csv('Fake.csv')
```

```python
true_news.head()
```

```python
fake_news.head()
```

```python
true_news['label'] = 1  # Real
fake_news['label'] = 0  # Fake
```

```python
true_news.head()
```

```python
fake_news.head()
```

```python
#Combine the two datasets
data = pd.concat([true_news, fake_news])
```

```python
data.head()
```

```python
data = data[['title' , 'text', 'label']]
```

```python
data.head(5)
```

```python
#Shuffle the dataset
data = data.sample(frac=1).reset_index(drop=True) 
```

```python
data['content'] = data['title'] + ' ' + data['text']
data.drop(['title', 'text'], axis=1, inplace=True)
```

```python
data.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(data['content'], data['label'], test_size=0.2, random_state=42)
```

```python
#Text Vectorization
tfidf = TfidfVectorizer(stop_words='english', max_df=0.7)
```

```python
tfidf
```

```python
X_train_vec = tfidf.fit_transform(X_train)
X_test_vec = tfidf.transform(X_test)
```

```python
#Logistic regression model
lr_model = LogisticRegression()
lr_model.fit(X_train_vec, y_train)
lr_preds = lr_model.predict(X_test_vec)
```

```python
# Naive Bayes model
nb_model = MultinomialNB()
nb_model.fit(X_train_vec, y_train)
nb_preds = nb_model.predict(X_test_vec)
```

```python
print("Logistic Regression Report:")
print(classification_report(y_test, lr_preds))
```

```python
print("Naive Bayes Report:")
print(classification_report(y_test, nb_preds))
```

```python
print("Logistic Regression Confusion Matrix:")
print(confusion_matrix(y_test, lr_preds))
```

```python
print("Naive Bayes Confusion Matrix:")
print(confusion_matrix(y_test, nb_preds))
```

```python
#Test
user_input = input("Enter the news content: ")
input_vec = tfidf.transform([user_input])
prediction = nb_model.predict(input_vec)[0]
if prediction == 1:
    print("The news is real.")
else:
    print("The news is fake.")
```
